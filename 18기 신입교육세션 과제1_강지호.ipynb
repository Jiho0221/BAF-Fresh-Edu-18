# 18기 신입교육세션 과제1_강지호
## 데이터명 : 보건복지부_의료서비스 이용률
 - 데이터 특징
 - 출처 : https://www.data.go.kr/data/15128032/fileData.do
 - 분야 : 의료
 - 내용 : 보건복지부에서 2017년부터 2023년까지 의료기관에서 의료서비스를 이용하는 이용객들의 이용률 현황(성별, 연령별, 거주지역별)에 대해 데이터 정보가 있습니다. 각 나이별, 성별, 지역별로 분류가 되어있어 연령별로 의료 이용 현황을 파악할 수 있습니다.
 - 건강보험 청구 데이터를 기반이므로 신뢰성이 있고 국내 의료 이용 패턴 분석에 용이합니다.

## 데이터명 : Healthcare Utilization by Age
 - 데이터특징
 - 출처 : https://www.kaggle.com/datasets/aranyogeshm/healthcare-dataset
 - 분야 : 의료
 - 내용 : kaggle에서 찾은 데이터로, 미국의 동북, 중서부, 남부, 서부지역들을 조사한 결과이며, 인구 정보, 동반 질환, 의료 이용률, 비용 등 다양한 의료 데이터가 있습니다.
 - 인구 통계와 성별, 지역, 건강 플랜 유형, 기본 지급자 유형이랑 질환 정보, 의료 이용 및 비용 등 다양한 데이터들이 많아 분석해볼 내용이 많다고 생각합니다.
 - 비용 예측이나 질병 위험 평가, 환자 순응도 모델링 등 의료 데이터 과학 응용에 유용합니다.
